## **Variables categoricas**
Las variables categóricas en un dataset son aquellas que contienen valores que representan categorías o grupos en lugar de números que tengan un significado matemático. Estas variables se utilizan para clasificar los datos en distintos grupos o etiquetas. Pueden ser tanto nominales como ordinales, dependiendo de si tienen o no un orden implícito

Variables Categóricas Nominales
Las variables categóricas nominales son aquellas que representan categorías sin un orden específico. Los valores no tienen una jerarquía, y solo se utilizan para clasificar o diferenciar.

Ejemplos:

Colores: rojo, azul, verde, amarillo.
Género: masculino, femenino.
País de origen: México, Perú, Argentina, Brasil.
Tipo de mascota: perro, gato, ave, pez.
Estilo de música favorito: rock, pop, jazz, clásico.

Variables Categóricas Ordinales
Las variables categóricas ordinales (o cardinales) tienen un orden o jerarquía. Los valores de estas variables se pueden clasificar en un rango, aunque las distancias entre ellos no son uniformes ni cuantificables.

Ejemplos:

Nivel educativo: primaria, secundaria, licenciatura, maestría, doctorado.
Satisfacción del cliente: muy insatisfecho, insatisfecho, neutral, satisfecho, muy satisfecho.
Tamaño de ropa: pequeño (S), mediano (M), grande (L), extra grande (XL).
Calificación de calidad: baja, media, alta.
Puntaje de dolor: leve, moderado, intenso.

En resumen, la diferencia principal es que las variables nominales no tienen un orden implícito, mientras que las ordinales sí.

### **Gráficas de frecuencias de las variables categóricas**

```python
# Definir cuántas columnas por fila tendrá la cuadrícula de gráficos
columns_per_row = 2
num_columns = len(var_categoricas)
num_rows = math.ceil(num_columns / columns_per_row)

# Crear la figura con subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=columns_per_row, figsize=(15, 6 * num_rows))
# Aplanar los ejes en caso de que haya más de una fila
axes = axes.flatten()

# Definir el color azul oscuro
color = 'darkblue'

# Iterar sobre las columnas y los ejes para graficar
for i, (ax, x) in enumerate(zip(axes, var_categoricas)):
    # Contar valores y calcular los porcentajes
    # Instead of using fillna directly on the DataFrame,
    # fill NaN values within the value_counts Series
    counts = df[x].value_counts().fillna("--NULL")
    total = len(df)
    # Crear el gráfico de barras con color azul oscuro
    counts.plot(kind='bar', ax=ax, title=x, color=color)
    # Añadir porcentaje sobre cada barra
    for p in ax.patches:
        percentage = f'{100 * p.get_height() / total:.2f}%'
        ax.annotate(percentage,
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center',
                    xytext=(0, 10),
                    textcoords='offset points')

# Si hay más ejes que columnas, ocultar los ejes vacíos
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
# Ajustar el layout para que los gráficos no se solapen
plt.tight_layout()
# Mostrar los gráficos
plt.show()
```

### **Crear otra variable numérica usando fuentes externas**
(Solo para curiosos)

Precio en Dolares

```python
# Función que convierte euros a dolares usando una API externa y usando caché
import requests
import time

cache = {
    "rate": None,
    "timestamp": None
}

def get_usd_to_eur_exchange_rate():
    global cache
    url = "https://api.exchangerate-api.com/v4/latest/USD"
    try:
        # Use cached rate if it's less than an hour old
        if cache["rate"] and (time.time() - cache["timestamp"] < 3600):
            print("Using cached exchange rate.")
            return cache["rate"]

        # Fetch a new rate from the API
        response = requests.get(url)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx and 5xx)
        data = response.json()

        eur_rate = data['rates'].get('EUR')
        if eur_rate:
            print(f"1 USD = {eur_rate} EUR")
            cache["rate"] = eur_rate
            cache["timestamp"] = time.time()
            return eur_rate
        else:
            print("EUR rate not found in the response.")
            return None

    except requests.exceptions.RequestException as e:
        print(f"Error fetching exchange rate: {e}")
        return None

def convert_eur_to_usd(amount_in_eur):
    exchange_rate = get_usd_to_eur_exchange_rate()
    if exchange_rate:
        amount_in_usd = amount_in_eur / exchange_rate
        print(f"{amount_in_eur} EUR = {amount_in_usd:.2f} USD")
        return amount_in_usd
    else:
        print("Conversion failed due to missing exchange rate.")
        return None

if __name__ == "__main__":
    euros = 100  # Example amount in EUR
    convert_eur_to_usd(euros)
```

```python
# Crear la variable Precio_USD
df['Precio_USD'] = 0.0  # Initialize a new column for USD prices

for index, row in df.iterrows():
    precio_eur = row['Precio']
    precio_usd = convert_eur_to_usd(precio_eur)
    if precio_usd is not None:
        df.loc[index, 'Precio_USD'] = precio_usd
    else:
        # Handle cases where conversion fails (e.g., set to NaN)
        df.loc[index, 'Precio_USD'] = float('nan')

print("Conversión finalizada")
```

### **3.1 Gráfico de Dispersión**

##### ¿Qué es y para qué sirve?

Un **gráfico de dispersión** es una herramienta que permite visualizar la relación entre dos variables cuantitativas. Los puntos en el gráfico representan la intersección de los valores de ambas variables para cada observación en el conjunto de datos.

El gráfico de dispersión es útil para:

1. **Analizar la relación causa-efecto** entre dos variables.
2. **Verificar hipótesis** sobre la relación entre las variables.
3. **Identificar patrones** en los datos. Si los puntos forman una banda que se extiende desde la parte inferior izquierda a la superior derecha, lo más probable es que exista una correlación positiva entre las dos variables. Si la banda va de la parte superior izquierda a la inferior derecha, es probable que exista una correlación negativa.
4. **Detectar valores atípicos**.
5. **Visualizar la correlación** entre las variables. Si los puntos de datos hacen una línea desde el origen desde los valores bajos de “x” y “y” hasta los valores altos de “x” y “y”, los puntos de datos están correlacionados positivamente. Si el gráfico comienza con valores altos de “y” y continúa con valores bajos de “y”, entonces el gráfico está correlacionado negativamente¹.

En resumen, un gráfico de dispersión es una herramienta poderosa para entender la relación entre dos variables y para explorar y analizar conjuntos de datos.

Crear dataset X con las variables Ram y Precio

```python
from sklearn.cluster import KMeans

# Selección de Variables: Se seleccionan las variables rm y medv para realizar el clustering,
# almacenándolas en la variable X.

X = df2[['RAM en GB', 'Precio']]
```

###1.1 Elección del Número de Clusters

Aplicaremos dos métodos: Método del Codo de Jambú (Elbow Method) y el Índice de Silhouette (Silhouette Score). Estos métodos son fundamentales para identificar la cantidad adecuada de clusters al aplicar algoritmos de agrupamiento como K-Means.

#### Método del Codo de Jambú
El Método del Codo ayuda a determinar el número óptimo de clusters observando la inercia (suma de las distancias cuadradas dentro de cada cluster) en función del número de clusters k.

```python
from sklearn.cluster import KMeans

# Selección de Variables: Se seleccionan las variables rm y medv para realizar el clustering,
# almacenándolas en la variable X.

X = df2[['RAM en GB', 'Precio']]

inertia = [] #Se crea una lista vacía inertia para almacenar la inercia para diferentes valores de k.
for k in range(1, 11):  # Probar con k desde 1 a 10

    kmeans = KMeans(n_clusters=k, random_state=0) #Instanciación del modelo: Para cada k, se crea un modelo KMeans con n_clusters=k y
                                                  #random_state=0 garantiza que los resultados sean reproducibles cada vez que se ejecute el código
    kmeans.fit(X) #Entrenamiento del modelo: Se ajusta el modelo a los datos X.
    inertia.append(kmeans.inertia_) #Se añade la inercia del modelo a la lista inertia

# Gráfico: Se busca el "codo" en la curva, donde la tasa de disminución de la inercia cambia significativamente.
#Este punto sugiere un buen número de clusters, ya que agregar más clusters no mejora significativamente la inercia.
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Método del Codo')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inertia')
plt.show()
```

#####Conclusiones

Muestra una gran disminución hasta k=3 ó 4 y luego la curva se “aplana”, lo que suele interpretarse como que a partir de ese punto los beneficios de agregar más clústeres son cada vez menores.

El método del Codo dentifica el punto donde la disminución de la inercia se estabiliza, sugiriendo que agregar más clusters no aporta una mejora significativa.
Por lo tanto, 3 ó 4 pueden ser valores adecuados

####Método del Índice de Silhouette

El Índice de Silhouette mide qué tan similar es cada punto a su propio cluster en comparación con otros clusters. Los valores del índice van de -1 a 1, donde valores más altos indican una mejor definición de los clusters.

```python
from sklearn.metrics import silhouette_score #evalúa la calidad de los cluster

silhouette_scores = []

for k in range(2, 11): # Probra k desde 2 to 10 (el índice de silhouette requiere al menos 2 clusters)
    kmeans = KMeans(n_clusters=k, random_state=0) #similar al método del codo
    kmeans.fit(X)
    silhouette_avg = silhouette_score(X, kmeans.labels_) # Cálculo del índice
                                                        #Se calcula el promedio del índice de Silhouette para todos los puntos usando silhouette_score.
    silhouette_scores.append(silhouette_avg) #Almacenamiento del índice: Se añade el valor promedio al listado silhouette_scores

#Visualización: Se grafica el índice de Silhouette en función de k.
plt.figure(figsize=(8, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Indice de Silhouette para maximizar k')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Indice de Silhouette')
plt.show()
```

#####Conclusiones

Para k=2 ves el valor más alto (por encima de 0.60), lo cual sugiere que dividir los datos en 2 clústeres produce la separación más nítida; sin embargo, hay un ligero “pico” secundario en 4.
A medida que aumentas el número de clústeres, el índice de Silhouette desciende de forma drástica y luego oscila (entre ~0.53 y ~0.57). Esto indica que, con más grupos, los datos comienzan a solaparse o a no separarse tan claramente.
**Insigth**: Significa que con solo dos grupos (posiblemente uno que engloba las laptops de menor RAM y precio más bajo y otro con mayor RAM y precio más alto) se obtiene la mejor separación “natural” según la métrica de Silhouette. Si nuestro objetivo es maximizar este índice, la recomendación sería escoger k=2. Sin embargo, la decisión final puede depender también de otros criterios (por ejemplo, de si tener solo 2 clústeres es útil desde el punto de vista del negocio).

Analizados ambos índices:
**bold text**
En la práctica, k=2 obtiene la mayor “pureza” de grupos (según Silhouette), pero puede resultar demasiado amplio (solo “grupo barato” vs. “grupo caro”).
Si deseamos un compromiso razonable entre buena separación y mayor detalle en los clústeres, k=4 suele ser un buen candidato:

El método del Codo sugiere que a partir de 3 o 4 la reducción de inercia deja de ser tan significativa.
El índice de Silhouette sigue siendo relativamente alto en 4, comparado con valores mayores de k.

**Por lo tanto, si tuviéramos que elegir un solo valor de k uniendo ambos criterios, 4 es un buen punto de equilibrio. **De todos modos, la decisión final también puede depender de la utilidad que cada número de clústeres tenga en tu contexto de negocio.
